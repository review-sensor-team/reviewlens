```python
import os
import google.generativeai as genai
from openai import OpenAI



def check_google_models():
    print("\nğŸ”µ [Google Gemini] ì‚¬ìš© ê°€ëŠ¥ ëª¨ë¸ ëª©ë¡ ì¡°íšŒ ì¤‘...")
    
    api_key = ''
    if not api_key:
        print("âŒ Google API Keyê°€ ì—†ìŠµë‹ˆë‹¤.")
        return

    try:
        genai.configure(api_key=api_key)
        
        # list_models()ë¥¼ í˜¸ì¶œí•˜ì—¬ ëª¨ë¸ ëª©ë¡ì„ ê°€ì ¸ì˜µë‹ˆë‹¤.
        # 'generateContent' ë©”ì„œë“œë¥¼ ì§€ì›í•˜ëŠ” ëª¨ë¸ë§Œ í•„í„°ë§í•©ë‹ˆë‹¤ (ì„ë² ë”© ëª¨ë¸ ì œì™¸).
        models = [
            m.name.replace('models/', '') # 'models/gemini-pro' -> 'gemini-pro'ë¡œ ë³´ê¸° ì¢‹ê²Œ ë‹¤ë“¬ê¸°
            for m in genai.list_models()
            if 'generateContent' in m.supported_generation_methods
        ]
        
        for model in sorted(models):
            print(f" - {model}")
            
    except Exception as e:
        print(f"âŒ ì¡°íšŒ ì‹¤íŒ¨: {e}")

def check_openai_models():
    print("\nğŸŸ¢ [OpenAI GPT] ì‚¬ìš© ê°€ëŠ¥ ëª¨ë¸ ëª©ë¡ ì¡°íšŒ ì¤‘...")
    
    api_key = ""
    if not api_key:
        print("âŒ OpenAI API Keyê°€ ì—†ìŠµë‹ˆë‹¤.")
        return

    try:
        client = OpenAI(api_key=api_key)
        
        # client.models.list()ë¥¼ í˜¸ì¶œí•©ë‹ˆë‹¤.
        model_list = client.models.list()
        
        # OpenAIëŠ” ëª¨ë¸ì´ ìˆ˜ë°± ê°œ(DALL-E, Whisper, êµ¬ë²„ì „ ë“±)ê°€ ë‚˜ì˜¤ë¯€ë¡œ 'gpt'ê°€ í¬í•¨ëœ ê²ƒë§Œ í•„í„°ë§í•©ë‹ˆë‹¤.
        gpt_models = [
            m.id 
            for m in model_list.data 
            if 'gpt' in m.id
        ]
        
        for model in sorted(gpt_models):
            print(f" - {model}")

    except Exception as e:
        print(f"âŒ ì¡°íšŒ ì‹¤íŒ¨: {e}")

if __name__ == "__main__":
    # check_google_models()
    check_openai_models()
```
